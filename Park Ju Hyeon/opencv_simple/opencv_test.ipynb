# IMPORT 

```python
import cv2
import sys
```

# FOR ONE IMAGE

```python
image_file = "./images/0.jpg"

cascade_file = "./haarcascade/haarcascade_frontalface_alt.xml"

image= cv2.imread(image_file)

image_gs = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

cascade = cv2.CascadeClassifier(cascade_file)

face_list = cascade.detectMultiScale(image_gs, scaleFactor=1.1, minNeighbors=1, minSize=(10, 10))

if len(face_list) > 0:
    print(face_list)
    color = (0,0,255)
    for face in face_list:
        x,y,w,h = face
        cv2.rectangle(image, (x,y), (x+w, y+h), color, thickness=2)
        
        cv2.imwrite("facedetect-output.PNG", image)
else:
    print("no face")
```

# FOR FIFTY IMAGES

```python
file_name = ".jpg"

cascade_file = "./haarcascade/haarcascade_frontalface_alt.xml"


for i in range(50) :
    image_file = "./person/" + str(i) + file_name
    image= cv2.imread(image_file)

    image_gs = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    cascade = cv2.CascadeClassifier(cascade_file)

    face_list = cascade.detectMultiScale(image_gs, scaleFactor=1.1, minNeighbors=1, minSize=(10, 10))

    if len(face_list) > 0:
        print(face_list)
        color = (0,0,255)
        for face in face_list:
            x,y,w,h = face
            cv2.rectangle(image, (x,y), (x+w, y+h), color, thickness=2)
            result_file = "./result/person" + str(i) + file_name
            cv2.imwrite(result_file, image)
    else:
        print("no face")
```
